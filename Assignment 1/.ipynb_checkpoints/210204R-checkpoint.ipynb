```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def transform_intensity(intensity):
    if intensity < 50:
        return intensity  # Linear part: y = x
    elif 50 <= intensity < 150:
        return 1.5 * intensity - 25  # Steep linear increase: y = 1.5x - 25
    elif 150 <= intensity < 200:
        return -2 * intensity + 500  # Linear decrease: y = -2x + 500
    else:
        return 255  # Saturate at 255

# Load the image in grayscale
image = cv2.imread('emma.jpg', cv2.IMREAD_GRAYSCALE)

# Vectorize the transformation function for use on numpy arrays
vectorized_transform = np.vectorize(transform_intensity)

# Apply the transformation
transformed_image = vectorized_transform(image.astype(np.float32)).astype(np.uint8)

# Plotting the original and transformed images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(transformed_image, cmap='gray')
plt.title('Transformed Image')
plt.axis('off')

plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Define the intensity transformation function:
# 0 to 50: The output intensity increases linearly from 0 to 50.
# 50 to 150: The output intensity increases linearly from 100 to 255.
# 150 to 255: The output intensity increases linearly from 150 to 255.

c = np.array([(50, 50), (150, 255)])  # Control points

# Create transformation segments
t1 = np.linspace(0, c[0, 1], c[0, 0] + 1).astype(np.uint8)  # 0 to 50 (0 -> 50)
t2 = np.linspace(c[0, 1] + 1, c[1, 1], c[1, 0] - c[0, 0]).astype(np.uint8)  # 50 to 150 (100 -> 255)
t3 = np.linspace(c[1, 1] + 1, 255, 255 - c[1, 0]).astype(np.uint8)  # 150 to 255 (150 -> 255)

# Combine all segments into one transformation array
transform = np.concatenate((t1, t2[1:], t3[1:]), axis=0).astype(np.uint8)

# Load the image (grayscale)
img_orig = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE)
assert img_orig is not None, "Image not found!"

# Apply the intensity transformation using cv.LUT
img_transformed = cv.LUT(img_orig, transform)

# Display the original and transformed images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(img_orig, cmap='gray')
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(img_transformed, cmap='gray')
ax[1].set_title('Transformed')
ax[1].axis('off')

plt.show()

# Optionally, save the transformed image
cv.imwrite('transformed_image.png', img_transformed)
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Define the intensity transformation function:
# 0 to 50: The output intensity increases linearly from 0 to 50.
# 50 to 150: The output intensity increases linearly from 100 to 255.
# 150 to 255: The output intensity increases linearly from 150 to 255.

c = np.array([(50, 50), (150, 255)])  # Control points

# Create transformation segments
t1 = np.linspace(0, c[0, 1], c[0, 0] + 1).astype(np.uint8)  # 0 to 50 (0 -> 50)
t2 = np.linspace(c[0, 1] + 1, c[1, 1], c[1, 0] - c[0, 0]).astype(np.uint8)  # 50 to 150 (100 -> 255)
t3 = np.linspace(c[1, 1] + 1, 255, 255 - c[1, 0]).astype(np.uint8)  # 150 to 255 (150 -> 255)

# Combine all segments into one transformation array
transform = np.concatenate((t1, t2[1:], t3[1:]), axis=0).astype(np.uint8)

# Ensure transform array has 256 elements and reshape to (256, 1)
transform = transform.reshape((256, 1)).astype(np.uint8)

# Load the image (grayscale)
img_orig = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE)
assert img_orig is not None, "Image not found!"

# Ensure the image is of type uint8
img_orig = np.uint8(img_orig)

# Apply the intensity transformation using cv.LUT
img_transformed = cv.LUT(img_orig, transform)

# Display the original and transformed images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(img_orig, cmap='gray')
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(img_transformed, cmap='gray')
ax[1].set_title('Transformed')
ax[1].axis('off')

plt.show()

# Optionally, save the transformed image
cv.imwrite('transformed_image.png', img_transformed)
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the image in grayscale
f = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE)
assert f is not None, "Error: Image not found!"

# Print the shape and dtype to verify the image is correctly loaded
print(f"Image shape: {f.shape}")
print(f"Image dtype: {f.dtype}")

# Ensure the image is a valid NumPy array and type uint8
f = np.array(f, dtype=np.uint8)
print(f"Processed Image dtype: {f.dtype}")  # This should be uint8

# Create a reverse intensity transformation array (255 -> 0)
t = np.arange(255, -1, -1, dtype=np.uint8).reshape((256, 1))

# Print the shape and dtype of LUT array
print(f"LUT shape: {t.shape}")
print(f"LUT dtype: {t.dtype}")

# Apply the intensity transformation using cv.LUT
g = cv.LUT(f, t)

# Display the original and transformed images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(f, cmap='gray')
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(g, cmap='gray')
ax[1].set_title('Transformed')
ax[1].axis('off')

plt.show()

# Optionally, save the transformed image
cv.imwrite('transformed_image.png', g)
```
